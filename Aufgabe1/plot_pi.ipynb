```python
import matplotlib.pyplot as plt
import pandas as pd

df_first = pd.read_csv('Pidata/memory_usage_1716725811.csv')
df_second = pd.read_csv('Pidata/memory_usage_1716726040.csv')
df_third = pd.read_csv('Pidata/memory_usage_1716726262.csv')

# memory usage is in bytes, convert to mb
df_first['Memory Usage [PSUtil] (bytes)'] = df_first['Memory Usage [PSUtil] (bytes)'] / 1000000
df_second['Memory Usage [PSUtil] (bytes)'] = df_second['Memory Usage [PSUtil] (bytes)'] / 1000000
df_third['Memory Usage [PSUtil] (bytes)'] = df_third['Memory Usage [PSUtil] (bytes)'] / 1000000

plt.plot(df_first['Time (s)'], df_first['Memory Usage [PSUtil] (bytes)'], label='First Run')
plt.plot(df_second['Time (s)'], df_second['Memory Usage [PSUtil] (bytes)'], label='Second Run')
plt.plot(df_third['Time (s)'], df_third['Memory Usage [PSUtil] (bytes)'], label='Third Run')

plt.xlabel('Time (s)')
plt.ylabel('Memory Usage (MB)')
plt.title('Memory Usage Over Time (PSUtil)')
plt.legend()
plt.show()
```

```python
# peak memory usage
m = [df_first['Memory Usage [PSUtil] (bytes)'].max(), df_second['Memory Usage [PSUtil] (bytes)'].max(), df_third['Memory Usage [PSUtil] (bytes)'].max()]
peak = sum(m) / len(m)
print(f'Peak memory usage: {peak} MB')
m
```

```python
# starting usage
starts = [df_first['Memory Usage [PSUtil] (bytes)'][0], df_second['Memory Usage [PSUtil] (bytes)'][0], df_third['Memory Usage [PSUtil] (bytes)'][0]]
start = sum(starts) / len(starts)
print(f'Starting memory usage: {start} MB')
```
