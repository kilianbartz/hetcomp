```python
import matplotlib.pyplot as plt
import pandas as pd

df_first = pd.read_csv('Serverdata/memory_usage_1716674587.csv')
df_second = pd.read_csv('Serverdata/memory_usage_1716676073.csv')
df_third = pd.read_csv('Serverdata/memory_usage_1716677535.csv')

# memory usage is in bytes, convert to mb
df_first['Memory Usage [PSUtil] (bytes)'] = df_first['Memory Usage [PSUtil] (bytes)'] / 1000000
df_second['Memory Usage [PSUtil] (bytes)'] = df_second['Memory Usage [PSUtil] (bytes)'] / 1000000
df_third['Memory Usage [PSUtil] (bytes)'] = df_third['Memory Usage [PSUtil] (bytes)'] / 1000000

plt.plot(df_first['Time (s)'], df_first['Memory Usage [PSUtil] (bytes)'], label='First Run')
plt.plot(df_second['Time (s)'], df_second['Memory Usage [PSUtil] (bytes)'], label='Second Run')
plt.plot(df_third['Time (s)'], df_third['Memory Usage [PSUtil] (bytes)'], label='Third Run')

plt.xlabel('Time')
plt.ylabel('Memory Usage')
plt.title('Memory Usage Over Time (PSUtil)')
plt.legend()
plt.show()
```

```python
# memory usage is in bytes, convert to mb
df_first['Current Memory Usage [Tracemalloc] (bytes)'] = df_first['Current Memory Usage [Tracemalloc] (bytes)'] / 1000000
df_second['Current Memory Usage [Tracemalloc] (bytes)'] = df_second['Current Memory Usage [Tracemalloc] (bytes)'] / 1000000
df_third['Current Memory Usage [Tracemalloc] (bytes)'] = df_third['Current Memory Usage [Tracemalloc] (bytes)'] / 1000000

plt.plot(df_first['Time (s)'], df_first['Current Memory Usage [Tracemalloc] (bytes)'], label='First Run')
plt.plot(df_second['Time (s)'], df_second['Current Memory Usage [Tracemalloc] (bytes)'], label='Second Run')
plt.plot(df_third['Time (s)'], df_third['Current Memory Usage [Tracemalloc] (bytes)'], label='Third Run')

plt.xlabel('Time')
plt.ylabel('Memory Usage')
plt.title('Memory Usage Over Time (tracemalloc)')
plt.legend()
plt.show()
```

```python
# memory usage is in bytes, convert to mb
df_first['Peak Memory Usage [Tracemalloc] (bytes)'] = df_first['Peak Memory Usage [Tracemalloc] (bytes)'] / 1000000
df_second['Peak Memory Usage [Tracemalloc] (bytes)'] = df_second['Peak Memory Usage [Tracemalloc] (bytes)'] / 1000000
df_third['Peak Memory Usage [Tracemalloc] (bytes)'] = df_third['Peak Memory Usage [Tracemalloc] (bytes)'] / 1000000

plt.plot(df_first['Time (s)'], df_first['Peak Memory Usage [Tracemalloc] (bytes)'], label='First Run')
plt.plot(df_second['Time (s)'], df_second['Peak Memory Usage [Tracemalloc] (bytes)'], label='Second Run')
plt.plot(df_third['Time (s)'], df_third['Peak Memory Usage [Tracemalloc] (bytes)'], label='Third Run')

plt.xlabel('Time')
plt.ylabel('Memory Usage')
plt.title('Peak Memory Usage Over Time (tracemalloc)')
plt.legend()
plt.show()
```

```python
# peak memory usage
peak = max(df_first['Peak Memory Usage [Tracemalloc] (bytes)'].max(), df_second['Peak Memory Usage [Tracemalloc] (bytes)'].max(), df_third['Peak Memory Usage [Tracemalloc] (bytes)'].max())
print(f'Peak memory usage: {peak} MB')
```
